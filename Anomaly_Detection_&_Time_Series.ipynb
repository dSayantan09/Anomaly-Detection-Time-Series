1. What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.

Ans: Anomaly Detection (also called outlier detection) is a data analysis technique used to identify observations, patterns, or events that deviate significantly from what is considered normal behavior in a dataset. Such deviations often indicate errors, rare events, fraud, faults, or emerging risks, depending on the domain.


Types of Anomalies
a. Point Anomalies

A point anomaly is a single data instance that is significantly different from the rest of the data.

Explanation:
The anomaly is evident when the data point is evaluated in isolation, without needing additional context.

Example:

In a credit card transaction dataset, a transaction of ‚Çπ5,00,000 when the user‚Äôs usual spending is ‚Çπ2,000‚Äì‚Çπ5,000.

In temperature sensor data, a sudden reading of 150¬∞C while all other readings are around 30¬∞C.

Key Characteristic:
One observation clearly stands out from the normal data distribution.

b. Contextual Anomalies

A contextual anomaly (also called conditional anomaly) is a data point that is anomalous only within a specific context, such as time, location, or environmental conditions.

Explanation:
The same value may be normal in one context but anomalous in another.

Example:

A temperature of 35¬∞C is normal during summer but anomalous during winter.

High website traffic at 12 PM may be normal, but the same traffic at 3 AM could be anomalous.

Key Characteristic:
Requires contextual information (time, season, location, etc.) to identify the anomaly.

c. Collective Anomalies

A collective anomaly occurs when a group of related data points, taken together, is anomalous, even though individual points may appear normal.

Explanation:
The anomaly is detected in the pattern or sequence rather than in individual values.

Example:

A sequence of small but frequent failed login attempts indicating a brute-force attack.

A gradual but sustained increase in network latency over time, signaling a system failure.

Key Characteristic:
Anomalous behavior emerges only when observing a collection or sequence of data points.

2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases.

Ans: a. Isolation Forest (IF)
Approach

Isolation Forest is a tree-based, unsupervised anomaly detection algorithm that isolates observations by randomly selecting features and split values.
Anomalies are isolated faster because they require fewer splits to be separated from the rest of the data.

Key Idea:
Anomalies are few and different ‚Üí they get isolated quickly.

Suitable Use Cases

High-dimensional datasets

Large-scale datasets (efficient and scalable)

Global anomaly detection

Fraud detection, network intrusion detection, system log monitoring

Strengths

No distance or density calculation

Works well with high dimensions

Low computational cost

Limitations

Less effective when anomalies are not clearly separable

Does not consider local data density

b. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
Approach

DBSCAN is a density-based clustering algorithm that groups points into clusters based on neighborhood density.
Points that do not belong to any dense region are labeled as noise (anomalies).

Key Idea:
Normal points form dense clusters; anomalies lie in sparse regions.

Suitable Use Cases

Spatial or low-dimensional data

Arbitrary-shaped clusters

Collective anomalies

Geospatial analysis, network traffic patterns, image segmentation

Strengths

No need to predefine number of clusters

Detects arbitrarily shaped clusters

Explicitly labels noise

Limitations

Sensitive to parameter selection (eps, min_samples)

Performs poorly in high-dimensional data

Struggles with varying density

c. Local Outlier Factor (LOF)
Approach

LOF is a density-based anomaly detection algorithm that measures how isolated a point is relative to its local neighborhood.
A point is anomalous if its local density is significantly lower than that of its neighbors.

Key Idea:
Anomalies have lower local density compared to surrounding points.

Suitable Use Cases

Local anomalies

Datasets with varying density

Context-dependent outliers

Fraud detection, sensor data analysis

Strengths

Captures local anomalies effectively

Works well with non-uniform density distributions

Limitations

Computationally expensive

Sensitive to choice of k (number of neighbors)

Not ideal for very large datasets

3. What are the key components of a Time Series? Explain each with one
example.

Ans: A time series is a sequence of observations recorded at regular time intervals (daily, monthly, yearly, etc.). Time series analysis focuses on understanding the underlying structure of these observations over time.

Key Components of a Time Series
1. Trend (T)

Definition:
The long-term movement or overall direction of the data over time.

Explanation:
Shows whether the series is increasing, decreasing, or remaining stable over a long period.

Example:
Annual growth in the number of internet users increasing steadily over several years.

2. Seasonal Component (S)

Definition:
Regular, repeating patterns that occur at fixed intervals (daily, weekly, monthly, or yearly).

Explanation:
Seasonality is caused by calendar-related effects such as weather, festivals, or business cycles.

Example:
Retail sales increasing every year during festive seasons like Diwali or Christmas.

3. Cyclical Component (C)

Definition:
Long-term fluctuations that occur over irregular periods, usually spanning multiple years.

Explanation:
Unlike seasonality, cycles do not have a fixed or predictable length.

Example:
Economic expansions and recessions affecting GDP over several years.

4. Irregular / Random Component (I)

Definition:
Unpredictable, random variations that cannot be explained by trend, seasonality, or cycles.

Explanation:
Caused by unexpected events such as natural disasters, strikes, or sudden policy changes.

Example:
A sudden drop in airline bookings due to an unexpected natural disaster.

4. Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?

Ans: A time series is said to be stationary if its statistical properties remain constant over time. Specifically:

The mean is constant.

The variance is constant.

The autocovariance (or autocorrelation) depends only on the lag, not on time.

Stationarity is a critical assumption for many time series models such as ARMA, ARIMA, and SARIMA.

How to Test Stationarity
a. Visual Inspection

Time series plot: Check for trend or changing variance.

Rolling mean and rolling variance: If these change over time, the series is likely non-stationary.

Example:
A sales series showing an upward trend indicates non-stationarity.

b. Statistical Tests
a) Augmented Dickey-Fuller (ADF) Test

Null hypothesis (H‚ÇÄ): Series is non-stationary (has a unit root).

Alternative hypothesis (H‚ÇÅ): Series is stationary.

Decision Rule:
If p-value < 0.05, reject H‚ÇÄ ‚Üí series is stationary.

b) KPSS Test

Null hypothesis (H‚ÇÄ): Series is stationary.

Alternative hypothesis (H‚ÇÅ): Series is non-stationary.

5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.

Ans: Autoregressive (AR) Model

The AR model explains the current value of a time series using its own past values. Its structure assumes that historical observations have a direct influence on the present observation.

From a structural perspective, an AR(p) model uses p lagged terms of the series itself. The model is suitable only when the time series is stationary.

In practice, AR models are applied when past values strongly influence future values, such as short-term forecasting of interest rates or demand where the series fluctuates around a stable mean.

Moving Average (MA) Model

The MA model represents the current value of a series as a function of past error terms (shocks) rather than past observations.

Structurally, an MA(q) model includes q lagged forecast errors. Like AR models, MA models also require the series to be stationary.

MA models are particularly useful when sudden random shocks have short-lived effects, for example in modeling sensor noise or correcting random fluctuations in signal processing.

ARIMA Model

ARIMA (Autoregressive Integrated Moving Average) extends AR and MA models by introducing differencing to handle non-stationary time series.

Structurally, ARIMA(p, d, q) combines:

AR terms (p),

differencing to achieve stationarity (d),

MA terms (q).

ARIMA is widely used for time series that show a trend but no strong seasonality, such as annual sales, economic indicators, or production metrics.

SARIMA Model

SARIMA (Seasonal ARIMA) further extends ARIMA to explicitly model seasonal patterns.

Its structure includes both non-seasonal components (p, d, q) and seasonal components (P, D, Q, m), where m represents the seasonal period (e.g., 12 for monthly data).

SARIMA is applied when the data exhibits regular, repeating seasonal behavior, such as monthly retail sales, airline passenger counts, or electricity consumption.

SARIMAX Model

SARIMAX (Seasonal ARIMA with eXogenous variables) builds on SARIMA by incorporating external explanatory variables that influence the time series.

Structurally, SARIMAX retains all SARIMA components and adds exogenous regressors (X), allowing the model to explain variation not captured by past values or seasonality alone.

SARIMAX is used in real-world forecasting scenarios where external factors matter, such as predicting sales using marketing spend, forecasting demand using weather data, or modeling economic output with policy indicators.

Conceptual Progression Summary

The models evolve in complexity as follows:
AR and MA handle stationary series without trend or seasonality; ARIMA adds differencing for trends; SARIMA adds seasonal structure; SARIMAX further incorporates external influencing variables.

This progression reflects increasing realism and applicability to complex, real-world time series data.

6.  Load a time series dataset (e.g., AirPassengers), plot the original series,
and decompose it into trend, seasonality, and residual components

```python
#Step 1: Load and Plot the Original Time Series
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load AirPassengers dataset
data = pd.read_csv(
    "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv",
    parse_dates=["Month"],
    index_col="Month"
)

# Plot original time series
plt.figure()
plt.plot(data, label="Air Passengers")
plt.title("AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

```python
#Step 2: Decompose the Time Series
# Decompose the time series
decomposition = seasonal_decompose(
    data["Passengers"],
    model="multiplicative",
    period=12
)

# Plot decomposition
decomposition.plot()
plt.show()
```

Output Interpretation
a. Trend Component

Shows a steady long-term increase in the number of passengers.

Captures overall growth in air travel demand.

b. Seasonal Component

Displays repeating yearly patterns.

Peaks correspond to high-travel months, and troughs correspond to low-demand months.

c. Residual Component

Represents random fluctuations not explained by trend or seasonality.

Mostly small, centered around 1 (due to multiplicative model).

7. Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.

```python
#Step 1: Load Dataset and Select Numerical Features
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# Load built-in taxi dataset
data = sns.load_dataset("taxis")

# Select numerical features
df = data[['distance', 'fare']].dropna()

# Rename for clarity
df.columns = ['trip_distance', 'fare_amount']
```

```python
#Step 2: Scale the Data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
#Step 3: Apply Isolation Forest
iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.03,
    random_state=42
)

df['anomaly'] = iso_forest.fit_predict(scaled_data)
# Convert labels: -1 ‚Üí anomaly, 1 ‚Üí normal
df['anomaly'] = df['anomaly'].map({1: 0, -1: 1})
```

```python
#Step 4: Visualize Anomalies in 2D
plt.figure()
plt.scatter(
    df[df['anomaly'] == 0]['trip_distance'],
    df[df['anomaly'] == 0]['fare_amount'],
    alpha=0.5,
    label='Normal'
)

plt.scatter(
    df[df['anomaly'] == 1]['trip_distance'],
    df[df['anomaly'] == 1]['fare_amount'],
    alpha=0.9,
    label='Anomaly'
)

plt.xlabel("Trip Distance")
plt.ylabel("Fare Amount")
plt.title("Isolation Forest Anomaly Detection (Taxi Fare Data)")
plt.legend()
plt.show()
```

8. Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.

```python
#Step 1: Load the Monthly Airline Passengers Dataset
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load AirPassengers dataset
data = pd.read_csv(
    "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv",
    parse_dates=["Month"],
    index_col="Month"
)

series = data["Passengers"]
```

```python
#Step 2: Train the SARIMA Model
# Fit SARIMA model
model = SARIMAX(
    series,
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

results = model.fit()
```

```python
#Step 3: Forecast the Next 12 Months
forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()
```

```python
#Step 4: Visualize the Forecast
plt.figure()
plt.plot(series, label="Observed")

plt.plot(forecast_mean, label="Forecast", linestyle="--")

plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    alpha=0.3,
    label="Confidence Interval"
)

plt.title("SARIMA Forecast: Airline Passengers (Next 12 Months)")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

9. Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib

```python
#Step 1: Load a Numerical Dataset
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.neighbors import LocalOutlierFactor
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Load dataset
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)

# Select two numerical features for 2D visualization
X = df[['alcohol', 'malic_acid']]
```

```python
#Step 2: Scale the Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Step 3: Apply Local Outlier Factor (LOF)
lof = LocalOutlierFactor(
    n_neighbors=20,
    contamination=0.05
)

# Fit and predict
labels = lof.fit_predict(X_scaled)

# Convert labels: -1 = anomaly, 1 = normal
df['anomaly'] = labels
```

```python
#Step 4: Visualize Anomalies (2D Scatter Plot)
plt.figure()

# Normal points
plt.scatter(
    X[df['anomaly'] == 1]['alcohol'],
    X[df['anomaly'] == 1]['malic_acid'],
    label='Normal',
    alpha=0.6
)

# Anomalies
plt.scatter(
    X[df['anomaly'] == -1]['alcohol'],
    X[df['anomaly'] == -1]['malic_acid'],
    label='Anomaly',
    alpha=0.9
)

plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.legend()
plt.show()
```

Ans: a. Real-Time Anomaly Detection Strategy
Choice of Algorithms

It detects global anomalies (sudden spikes/drops)

LOF is useful for detecting local anomalies within a specific region or substation where consumption patterns differ across regions.

DBSCAN (Not ideal for real-time streaming)
DBSCAN is less suitable because:

Sensitive to parameter tuning

Computationally expensive for streaming

Poor performance in high-dimensional, evolving data

b. Short-Term Forecasting Model Selection

Why SARIMAX over ARIMA / SARIMA

Energy demand shows strong daily and weekly seasonality

External factors (weather, region, holidays) significantly impact demand

SARIMAX allows exogenous variables like temperature and humidity

Use case

Short-term forecasting (next few hours to next day)

Real-time grid load planning

c. Validation and Continuous Monitoring
Forecast Validation

Rolling / walk-forward validation

Metrics:

MAE (Mean Absolute Error)

RMSE (Root Mean Squared Error)

MAPE (for business interpretability)

Anomaly Detection Monitoring

Track:

Anomaly rate per region

Sudden increase in anomaly frequency

Periodic retraining using sliding time windows

Model Drift Handling

Re-train forecasting model weekly/monthly

Monitor residuals for bias or variance shift

Alert when forecast error exceeds thresholds

d. Business & Operational Impact

This solution enables:

Early detection of grid failures or sensor faults

Preventive maintenance before outages

Load balancing and peak demand planning

Cost optimization by reducing emergency power purchases

Regulatory compliance through anomaly audits

```python
#Step1: Anomaly Detection Using Isolation Forest
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# Simulated energy usage data
data = pd.DataFrame({
    "energy_usage": [420, 430, 440, 1000, 450, 460, 200, 470, 480]
})

# Scale data
scaler = StandardScaler()
scaled = scaler.fit_transform(data)

# Train Isolation Forest
iso = IsolationForest(contamination=0.1, random_state=42)
data["anomaly"] = iso.fit_predict(scaled)

# -1 = anomaly, 1 = normal
print(data)
```

```python
#Step2: Short-Term Forecasting Using SARIMAX
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Simulated time series
series = pd.Series(
    [420, 430, 440, 450, 460, 470, 480],
    index=pd.date_range("2024-01-01", periods=7, freq="15min")
)

# SARIMAX model
model = SARIMAX(
    series,
    order=(1,1,1),
    seasonal_order=(1,0,1,96)  # 96 intervals = 1 day
)

results = model.fit()
forecast = results.forecast(steps=4)

print(forecast)
```
